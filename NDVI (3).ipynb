```python
# 1. Import libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold, cross_val_score, train_test_split
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings("ignore")

# 2. Load the data
df = pd.read_csv("hacktrain.csv")
print("Shape:", df.shape)

# 3. NDVI columns
ndvi_cols = [col for col in df.columns if col.endswith('_N')]

# 4. Handle missing values: interpolate + mean fill
df[ndvi_cols] = df[ndvi_cols].interpolate(axis=1, limit_direction='both')
df[ndvi_cols] = df[ndvi_cols].fillna(df[ndvi_cols].mean())

# 5. Outlier removal using z-score
from scipy.stats import zscore
z_scores = df[ndvi_cols].apply(zscore)
df[ndvi_cols] = df[ndvi_cols].mask(np.abs(z_scores) > 3)
df[ndvi_cols] = df[ndvi_cols].interpolate(axis=1, limit_direction='both')
df[ndvi_cols] = df[ndvi_cols].fillna(df[ndvi_cols].mean())

# 6. Standard scaling
scaler = StandardScaler()
df[ndvi_cols] = scaler.fit_transform(df[ndvi_cols])

# 7. Feature Engineering
df['ndvi_mean'] = df[ndvi_cols].mean(axis=1)
df['ndvi_std'] = df[ndvi_cols].std(axis=1)
df['ndvi_max'] = df[ndvi_cols].max(axis=1)
df['ndvi_min'] = df[ndvi_cols].min(axis=1)
df['ndvi_range'] = df['ndvi_max'] - df['ndvi_min']

# 8. Prepare features and labels
X = df[ndvi_cols + ['ndvi_mean', 'ndvi_std', 'ndvi_max', 'ndvi_min', 'ndvi_range']]
y = df['class']

# 9. Encode target if needed (Optional, depends on format)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# 10. Logistic Regression with class weights
model = LogisticRegression(
    multi_class='multinomial',
    solver='lbfgs',
    class_weight='balanced',
    max_iter=1000,
    random_state=42
)

# 11. Cross-validation to check performance
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scores = cross_val_score(model, X, y_encoded, cv=cv, scoring='accuracy')
print("Cross-validated Accuracy:", scores.mean())

# 12. Train-test split for actual training
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

# 13. Train the model
model.fit(X_train, y_train)

# 14. Evaluate on test set
y_pred = model.predict(X_test)
print("Test Accuracy:", accuracy_score(y_test, y_pred))
```
